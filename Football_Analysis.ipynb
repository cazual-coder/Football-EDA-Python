## Data Loading and Overview

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
epl_data=pd.read_csv("5_Football_Project.csv")

epl_data.head()
```

```python
# Show dataset structure and data types
epl_data.info()
```

```python
# Summary statistics for numeric columns
epl_data.describe()
```

```python
# Check for missing values
epl_data.isnull().sum()
```

```python
#create 2 new columns
```

```python
# Add average minutes and goals per match columns
epl_data['MinutesPerMatch']=(epl_data['Mins']/epl_data['Matches']).astype(int)
epl_data['GoalsPerMatch']=(epl_data['Goals']/epl_data['Matches']).astype(int)
epl_data
```

```python
# Adding average minutes and goals per match columns
epl_data['MinutesPerMatch'].dtype
```

```python
# Calculate average values (mean)
goal_mean=np.mean(epl_data['Goals'])
goal_mean
```

```python
# Calculate average values (mean)
assists_mean=np.mean(epl_data['Assists'])
assists_mean
```

## Club-Level Analysis

```python
## players based on age
```

```python
Under20=epl_data[epl_data['Age']<=20]
age21_25=epl_data[(epl_data['Age']>20)&(epl_data['Age']<=25)]
age25_30=epl_data[(epl_data['Age']>25)&(epl_data['Age']<=30)]
Above30=epl_data[epl_data['Age']>30]
```

```python
x=np.array([Under20['Name'].count(),age21_25['Name'].count(),age25_30['Name'].count(),Above30['Name'].count()])
```

```python
x
```

```python
# Pie chart visualization
plt.pie(x,labels= ["<=20","21-25","25-30",">=30"],autopct="%.1f%%")
plt.title('Total Players with Age groups',fontsize=20);
```

```python
# Count occurrences by category
## total players under 20 in each club

pl_u20=epl_data[epl_data['Age']<20]
pl_u20['Club'].value_counts().plot(kind='bar',color=sns.color_palette('cubehelix'))
```

## Age and Demographic Analysis

```python
# Calculate total values
plt.figure(figsize=(12,6))
Assists_by_clubs = pd.DataFrame(epl_data.groupby('Club',as_index=False)['Assists'].sum())
ax=sns.barplot(data=Assists_by_clubs.sort_values(by='Assists'),x="Club",y="Assists")
ax.set_xlabel('Clubs',fontsize=15)
ax.set_ylabel('Assists',fontsize=15)
plt.title("Clubs vs Assists",fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.show()
```

```python
# Calculate total values
plt.figure(figsize=(12,6))
Assists_by_clubs = pd.DataFrame(epl_data.groupby('Club',as_index=False)['Goals'].sum())
ax=sns.barplot(data=Assists_by_clubs.sort_values(by='Goals'),x="Club",y="Goals")
ax.set_xlabel('Clubs',fontsize=15)
ax.set_ylabel('Goals',fontsize=15)
plt.title("Clubs vs Goals",fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.show()
```
